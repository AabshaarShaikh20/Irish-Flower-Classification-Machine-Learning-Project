# <span style="color:purple">----- Let's build our model -----</span>


```python
#importing Library
from sklearn.model_selection import train_test_split
```

```python
#Seeing where our Target Var is..
irisflower.columns
```

```python
train, test = train_test_split(irisflower, test_size=.20)

# Separate features (X) and target variable (y) for training set
train_x = train.iloc[:, 0:-1]  # All columns except the first one as features
train_y = train.iloc[:, -1]   # The first column as target variable

# Separate features (X) and target variable (y) for testing set
test_x = test.iloc[:, 0:-1]    # All columns except the first one as features
test_y = test.iloc[:, -1]     # The first column as target variable
```

# <span style="color:#7D0541">----- Logestic Regression -----</span>



```python
#importing Library
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression 

logreg = LogisticRegression()                          #creating a function of it & Storing it in a variable named as 'logreg'
logreg.fit(train_x, train_y)                            #model is created

pred_logreg = logreg.predict(test_x)
print(confusion_matrix(test_y, pred_logreg))            
print(classification_report(test_y, pred_logreg)) 
```

```python
from sklearn.metrics import accuracy_score
```

```python
logreg_accuracy = accuracy_score(test_y, pred_logreg)
print("Logestic Regression Accuracy = ",logreg_accuracy)
```

# <span style="color:#7D0541">----- Decision Tree -----</span>



```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()

dt.fit(train_x, train_y)

pred_dt = dt.predict(test_x)
print(confusion_matrix(test_y, pred_dt))            
print(classification_report(test_y, pred_dt)) 
```

```python
dt_accuracy = accuracy_score(test_y, pred_dt)
print("Decision Tree Accuracy = ",dt_accuracy)
```

# <span style="color:#7D0541">----- Random Forest -----</span>



```python
# Experimenting with different algorithms, ensemble methods
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()

rf.fit(train_x, train_y) #model is build

# Evaluate the Random Forest model
pred_rf = rf.predict(test_x)
print(confusion_matrix(test_y, pred_rf))            
print(classification_report(test_y, pred_rf)) 
```

```python
rf_accuracy = accuracy_score(test_y, pred_rf)
print("Random Forest Accuracy = ",rf_accuracy)
```

# <span style="color:#7D0541">----- SVM -----</span>



```python
# Support vector machine algorithm
from sklearn.svm import SVC
svm = SVC()
svm.fit(train_x, train_y) #model is build

# Evaluating the model
pred_svm = svm.predict(test_x)
print(confusion_matrix(test_y, pred_svm))            
print(classification_report(test_y, pred_svm)) 
```

```python
svm_accuracy = accuracy_score(test_y, pred_svm)
print("Support Vector Machine Accuracy = ",svm_accuracy)
```

# <span style="color:#7D0541">----- KNN -----</span>



```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(train_x, train_y)

# Evaluating the model
pred_knn = knn.predict(test_x)
print(confusion_matrix(test_y, pred_knn))            
print(classification_report(test_y, pred_knn)) 
```

```python
knn_accuracy = accuracy_score(test_y, pred_knn)
print("KNN Accuracy = ",knn_accuracy)
```

# <span style="color:brown">Comparing All Results</span>


```python
print("Logestic Regression Accuracy = ", logreg_accuracy)
print("Decision Tree Accuracy       = ", dt_accuracy)
print("Random Forest Accuracy       = ", rf_accuracy)
print("Svm Accuracy                 = ", svm_accuracy)
print("KNN Accuracy                 = ", knn_accuracy)
```

```python
# MSE values for each method on test data
methods     = ['Logestic Regression', 'Decision Tree', 'Random Forest', 'SVM', 'KNN']
mse_values = [logreg_accuracy, dt_accuracy, rf_accuracy, svm_accuracy, knn_accuracy]

# Creating a bar plot
plt.figure(figsize=(10, 6))
plt.bar(methods, mse_values, color=['yellow', 'gold', 'orange', 'coral', 'lightcoral'])

# Adding labels and title
plt.xlabel('Models')
plt.ylabel('Accuracy Value')
plt.title('Comparing Accuracy for Different Methods')
plt.xticks(rotation=45)
plt.grid()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
'''

After experimenting with various algorithms, we can conclude that both Logestic Regression and Support Vector Machine(SVM),
delivered the best results for Irish Flower Classification. 

This model showed superior accuracy and performance across several evaluation metrics compared to others.

'''
```
